 1. What is the difference between a function and a method in Python?
  - Function : In Python, a function is a block of code that performs a specific task and can be defined independently using the def keyword. It is not tied to any object and can be called directly.
  example :
    def greet(name):
    return f"Hello, {name}!"
  Method : A method, on the other hand, is a function that is associated with an object typically defined inside a class. It automatically takes the instance (self) as its first argument, allowing it to access and modify the object's attributes.
  example :
    class Greeter:
    def greet(self, name):
        return f"Hello, {name}!"

 Keyword arguments: Matched by name.
 Default arguments: Parameters with default values.
 Variable-length arguments: Using *args and **kwargs to accept multiple values

 6. Explain the concept of generators in Python and how they are defined

 Defined like a regular function using def
 Uses the yield keyword instead of return
 Automatically creates an iterator object
 Program :
    def count_up_to(n):
      count = 1
      while count <= n:
        yield count
        count += 1

 7. What are the advantages of using generators over regular functions?
 - Generators offer several advantages over regular functions in Python, especially when working with large datasets or streams of data :    
 Key Advantages of Generators
 Memory Efficiency: Generators yield one item at a time, so they don't store the entire sequence in memory. This is ideal for handling large files or infinite sequences.
 Lazy Evaluation: Values are produced only when needed, which speeds up execution and reduces resource usage.
 Simplified Code: Using yield makes it easy to write clean, readable code for iterators without managing state manually.
 Infinite Sequences: You can create sequences that don't have a fixed end, like generating numbers forever, without crashing your program.
 Better Performance: For tasks like reading large logs or streaming data, generators are faster and more responsive than building full lists.
 Program :    
    def count_up_to(n):
      count = 1
      while count <= n:
        yield count
        count += 1

 Purpose of map()
  To simplify code by applying a function to each element in a collection.

```python
# 1. Write a Python function that takes a list of numbers as input and returns the sum of all even numbers in the list
def sum_even(numbers):
    total = 0
    for num in numbers:
        if num % 2 == 0:
            total += num
    return total

user_input = input("Enter numbers separated by spaces: ")
numbers = list(map(int, user_input.split()))

print("Sum of even numbers:", sum_even(numbers))
```

```python
# 4. Write a Python function that checks if a given number is prime or not from 1 to 200
def is_prime(n):
    if n < 2 or n > 200:
        return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True

num = int(input("Enter a number between 1 and 200: "))

if is_prime(num):
    print(f"{num} is a prime number.")
else:
    print(f"{num} is not a prime number.")
```

```python
# 5. Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of terms
class FibonacciIterator:
    def __init__(self, max_terms):
        self.max_terms = max_terms
        self.count = 0
        self.a = 0
        self.b = 1

    def __iter__(self):
        return self

    def __next__(self):
        if self.count >= self.max_terms:
            raise StopIteration
        if self.count == 0:
            self.count += 1
            return self.a
        elif self.count == 1:
            self.count += 1
            return self.b
        else:
            self.a, self.b = self.b, self.a + self.b
            self.count += 1
            return self.b

terms = int(input("Enter the number of Fibonacci terms: "))

fib = FibonacciIterator(terms)
for num in fib:
    print(num, end=" ")
```

```python
# 6. Write a generator function in Python that yields the powers of 2 up to a given exponent.
def power_of_two(exponent):
    for i in range(exponent + 1):
        yield 2 ** i

exponent = int(input("Enter the exponent: "))

for power in power_of_two(exponent):
    print(power)
```

```python
# 7. Implement a generator function that reads a file line by line and yields each line as a string
def read_file_lines(file_path):
    try:
        with open(file_path, 'r') as file:
            for line in file:
                yield line.strip()
    except FileNotFoundError:
        print(f"Error: File '{file_path}' not found.")

file_path = input("Enter the file path: ")

for line in read_file_lines(file_path):
    print(line)

```

```python
# 9. Write a Python program that uses `map()` to convert a list of temperatures from Celsius to Fahrenheit
def c_to_f(celsius):
    return round(celsius * 9/5 + 32, 2)

user_input = input("Enter Celsius temperatures separated by spaces: ")

celsius_temps = list(map(float, user_input.split()))

fahrenheit_temps = list(map(c_to_f, celsius_temps))

print("Celsius temperatures:", celsius_temps)
print("Fahrenheit temperatures:", fahrenheit_temps)
```

```python
# 10. Create a Python program that uses `filter()` to remove all the vowels from a given string
def remove_vowels(text):
    vowels = 'aeiouAEIOU'
    return ''.join(filter(lambda char: char not in vowels, text))

user_input = input("Enter a string: ")

result = remove_vowels(user_input)
print("String without vowels:", result)
```
